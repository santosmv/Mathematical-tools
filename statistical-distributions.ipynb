```python
import numpy as np
from scipy.special import factorial

def gaussian(x, mu, sigma):
    gauss = 1/np.sqrt(2 * np.pi * sigma**2) * np.exp(- (x - mu)**2 / (2 * sigma**2))
    return gauss

def binomial(n, N, p):
    binom = factorial(N) / (factorial(n) * factorial(N - n)) * p**n * (1-p)**(N-n)
    return binom

def poisson(n, N, p):
    nu = N*p
    pois = nu**n / factorial(n) * np.exp(-nu)
    return pois
```

Gaussian:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('style.mplstyle')
import warnings
warnings.filterwarnings("ignore")

plt.figure(dpi=160)

mu_list = [0, -5, 3]
sigma_list = [1, 2, 4]

x = np.linspace(-15, 15, 500)

for i in range(len(mu_list)):
    mu = mu_list[i]
    sigma = sigma_list[i]
    plt.plot(x, gaussian(x, mu, sigma), label=r'$\mu = %i$  $\sigma = %i$'%(mu, sigma))

markerline, stemlines, baseline = plt.stem(mu_list, gaussian(np.array(mu_list), np.array(mu_list), np.array(sigma_list)), '--', label='Expected value')
plt.setp(stemlines, 'color', 'gainsboro')
plt.setp(markerline, 'color', 'silver')
plt.setp(baseline, visible=False)

plt.xlabel('x', fontsize=15)
plt.ylabel(r'$P(x, \mu, \sigma)$', fontsize=15)
plt.xlim(-15,15)
plt.ylim(0,0.5)
plt.legend()
plt.show()
```

Binomial:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(dpi=160)

p_list = [0.1, 0.5, 0.1]
N_list = [10, 20, 40]

n = np.arange(0, 20)

for i in range(len(p_list)):
    p = p_list[i]
    N = N_list[i]
    plt.plot(n, binomial(n, N, p), 's', label=r'$p = %.1f$'%p + r'   $N = %i$'%N)

markerline, stemlines, baseline = plt.stem(np.array(N_list) * np.array(p_list), binomial(np.array(N_list) * np.array(p_list), np.array(N_list), np.array(p_list)), '--', label='Expected value')
plt.setp(stemlines, 'color', 'gainsboro')
plt.setp(markerline, visible=False)
plt.setp(baseline, visible=False)

plt.xlabel('n', fontsize=15)
plt.ylabel(r'$P(n, N, p)$', fontsize=15)
plt.xticks(range(0,20))
plt.xlim(0,16)
plt.ylim(0,0.5)
plt.legend()
plt.show()
```

Poisson:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(dpi=160)

p_list = np.array([0.9, 0.8, 0.05])
N_list = np.array([5, 10, 50])
nu_list = p_list * N_list

n = np.arange(0, 20)

for i in range(len(p_list)):
    p = p_list[i]
    N = N_list[i]
    plt.plot(n, poisson(n, N, p), 's', label=r'$\nu = %.1f$'%(N*p) + r'    $N = %i$'%N)

markerline, stemlines, baseline = plt.stem(np.array(N_list) * np.array(p_list), poisson(np.array(N_list) * np.array(p_list), np.array(N_list), np.array(p_list)), '--', label='Expected value', use_line_collection=True)
plt.setp(stemlines, 'color', 'gainsboro')
plt.setp(markerline, 'color', 'gainsboro')
plt.setp(baseline, visible=False)

plt.xlabel('n', fontsize=15)
plt.ylabel(r'$P(n,\nu)$', fontsize=15)
plt.xticks(range(0,20))
plt.xlim(0,16)
plt.ylim(0,0.3)
plt.legend()
plt.show()
```
