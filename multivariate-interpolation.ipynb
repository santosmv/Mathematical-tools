```python
import numpy as np
from scipy.interpolate import RegularGridInterpolator
def f(x, y, z):
    return 2 * x**3 + 3 * y**2 - z

n = 2
x = np.linspace(1, 2, n)
y = np.linspace(2, 3, n+1)
z = np.linspace(3, 4, n+2)
xg, yg ,zg = np.meshgrid(x, y, z, indexing='ij', sparse=True)
# data = f(xg, yg ,zg)
f_vec = np.vectorize(f)
data = f_vec(xg, yg ,zg)

print(data.shape)
print(data)
interp = RegularGridInterpolator((x, y, z), data)
a = interp([1,2.5,3.5])
# print(a.shape)
# print(a[0])
```

```python
import numpy as np
from scipy.interpolate import RegularGridInterpolator
def f(t,u,v,x, y, z):
    return 2 * x**3 + 3 * y**2 - z + 2 * t**3 + 3 * u**2 - v

n = 11
t = np.linspace(1, 4, n)
u = np.linspace(4, 7, n)
v = np.linspace(7, 9, n)
x = np.linspace(1, 4, n)
y = np.linspace(4, 7, n)
z = np.linspace(7, 9, n)
tg, ug ,vg, xg, yg ,zg = np.meshgrid(t, u, v, x, y, z, indexing='ij', sparse=True)
data = f(tg, ug ,vg, xg, yg ,zg)

print(data)
interp = RegularGridInterpolator((t, u, v, x, y, z), data)
a = interp([2,5,8,2,5,8])
a[0]
```
